## Training Data Loading / Preprocessing

```python
train_ds = CIFAR10(root, train=True)
train_ds.show_sample_images()
```

```python
X_train_raw = train_ds.images
y_train_raw = train_ds.labels

X_train_raw.shape, y_train_raw.shape
```

```python
train_samples = 30000

X_train = X_train_raw[:train_samples]
y_train = y_train_raw[:train_samples]

X_train.shape, y_train.shape
```

```python
def convert_gray(xtrain):
    xtrain_grayed = []
    for i in range(len(xtrain)):
        gray_image = cv2.cvtColor(xtrain[i], cv2.COLOR_RGB2GRAY)
        xtrain_grayed.append(gray_image)

    # Convert the list to a NumPy array
    xtrain_grayed = np.array(xtrain_grayed)

    return xtrain_grayed

# Example usage:
# Assuming 'xtrain' is the CIFAR dataset after splitting

# Perform grayscale conversion
xtrain_grayed = convert_gray(X_train)
```

```python
import cv2
import numpy as np

def apply_hog_descriptor(images_array):
    hog_features_list = []

    for i in range(images_array.shape[0]):
        image = images_array[i]

        # Convert the image to grayscale
        gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

        # Calculate HOG features
        features = hog(gray_image, orientations=9, pixels_per_cell=(8, 8),
                       cells_per_block=(3, 3))

        hog_features_list.append(features)

    # Convert the list of features to a NumPy array
    hog_features_array = np.array(hog_features_list)

    return hog_features_array

# Example usage:
# Assuming 'images_data' is the NumPy array with images of shape (5000, 32, 32, 3)

# Apply HOG descriptor to the array of images
X_hog = apply_hog_descriptor(X_train)

# Check the shape of the resulting HOG features array
print("Shape of HOG features:", X_hog.shape)
```

## Training

```python
n_classes = 10  # CIFAR-10 has 10 classes
model = OneVsRestSVM(n_classes)
model.fit(X_hog, y_train, kernel='hog')
```

```python
def test_svm(svm, X_test, y_test):
    # Make predictions
    predictions = svm.predict(X_test)
    print(max(predictions), min(predictions))
    # Calculate accuracy
    correct_predictions = np.sum(predictions == y_test)
    accuracy = correct_predictions.item() / y_test.shape[0]
    
    return accuracy
```
