```python
import pickle
def unpickle(file):
    with open(file, 'rb') as fo:
        data = pickle.load(fo, encoding='latin1')
    return data

def load_cifar10_batch(batch_file):
    batch = unpickle(batch_file)
    features = batch['data']
    labels = batch['labels']
    return features, labels
label_names = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]

```

```python
import matplotlib.pyplot as plt
import numpy as np

def show_sample_images_with_labels(features, labels, label_names, num_samples=5):
    fig, axs = plt.subplots(1, num_samples, figsize=(15, 3))
    for i in range(num_samples):
        img = features[i].reshape(3, 32, 32).transpose(1, 2, 0)
        axs[i].imshow(img)
        axs[i].set_title(label_names[labels[i]])
        axs[i].axis('off')
    plt.show()

# Example usage

batch1_features, batch1_labels = load_cifar10_batch('/scratch1/apicker/data/cifar10/train/data_batch_1')
batch2_features, batch2_labels = load_cifar10_batch('/scratch1/apicker/data/cifar10/train/data_batch_3')

# Assuming you have loaded the features and labels for each batch
print("Sample images from Batch 1:")
show_sample_images_with_labels(batch1_features, batch1_labels, label_names)

print("\n")
print("Sample images from Batch 2:")
show_sample_images_with_labels(batch2_features, batch2_labels, label_names)

```
